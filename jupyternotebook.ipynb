```python
"""
Single-use script to generate Polymarket CLOB API credentials.
Run once, save your credentials somewhere safe, then delete this file.

Prerequisites:
    pip install py-clob-client

Usage:
    python generate_polymarket_creds.py
"""

from py_clob_client.client import ClobClient

# ── CONFIG ──────────────────────────────────────────────────────────
PRIVATE_KEY = "0x5c1102573c871e6d64af9d8b043801f7860de5c233ba0a16c2ee500123b0d5db"  # Replace with your wallet private key
HOST = "https://clob.polymarket.com"
CHAIN_ID = 137  # Polygon mainnet
# ────────────────────────────────────────────────────────────────────

def main():
    client = ClobClient(HOST, key=PRIVATE_KEY, chain_id=CHAIN_ID)
    creds = client.create_or_derive_api_creds()

    print("\n✅  Polymarket L2 API Credentials")
    print("=" * 42)
    print(f"  API Key:    {creds.api_key}")
    print(f"  Secret:     {creds.api_secret}")
    print(f"  Passphrase: {creds.api_passphrase}")
    print("=" * 42)
    print("\n⚠️  Save these somewhere safe, then delete this script.\n")

if __name__ == "__main__":
    main()
```
